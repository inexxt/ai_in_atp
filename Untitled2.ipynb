```python
bads = [14, 13, 11, 10, 1, 8, 7, 5, 4, 2]
```

```python
with open("./data/flat30-100.cnf", "r") as f:
    ls = f.readlines()
```

```python
for t in range(2**len(bads)):
    with open(f"./mydata/flat{t:010b}", "w") as f:
        f.writelines(ls)
        for k in range(len(bads)):
            if t & 2**k:
                print("A", t, k)
                f.write(f"{bads[k]} 0\n")
            else:
                print("B", t, k)
                f.write(f"{-bads[k]} 0\n")
```
